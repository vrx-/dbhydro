```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
path = 'datasets/REPORT-8.csv'
df = pd.read_csv(path, parse_dates=['Collection_Date'])
nvars = df['Test Name'].unique()
nvars
```

```python
df.head()
```

```python
df.dtypes
```

```python
fig, axes = plt.subplots(2,1, figsize=(12,7), sharex=True)
ylims = ([.15, .4]) # upper limits for y axis
for var, ax, lim in zip(nvars[:-1], axes, ylims):
    sub = df.where(df['Test Name']==var)
    sub.plot(x='Collection_Date', y='Value', ax=ax, legend=False)
    ax.set_ylabel(var+' [mg/L]')
    ax.set_ylim(0, lim)
    
    ax2 = ax.twinx()
    var2 = sub['Value']/.031
    ax2.plot(sub['Collection_Date'], var2)
    ax2.set_ylabel(var+r' [$\mu$M]')
    ax2.grid(None)
    ax2.set_ylim(0, lim/.031)

# fig.suptitle('L001', )
# df.where(df['Test Name']==var).plot(x='Collection_Date', y='Value', ax=axes[0])
# df.where(df['Test Name']==var).plot(x='Collection_Date', y='Value', ax=axes[0])
```

```python
path = 'datasets/REPORT-9.csv'
df2 = pd.read_csv(path, parse_dates=['Collection_Date'])
nvars2 = df2['Test Name'].unique()
nvars2
```

```python
fig, axes = plt.subplots(2,1, figsize=(12,7), sharex=True)
ylims = ([.7, 3])
for var, ax, lim in zip(nvars2, axes, ylims):
    sub = df2.where(df2['Test Name']==var)
    sub.plot(x='Collection_Date', y='Value', ax=ax, legend=False)
    ax.set_ylabel(var+' [mg/L]')
    ax.set_ylim(0, lim)
    
    ax2 = ax.twinx()
    var2 = sub['Value']/.014
    ax2.plot(sub['Collection_Date'], var2)
    ax2.set_ylabel(var+r' [$\mu$M]')
    ax2.grid(None)
    ax2.set_ylim(0, lim/.014)
```
